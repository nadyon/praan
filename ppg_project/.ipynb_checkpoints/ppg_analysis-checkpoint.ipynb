```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import signal
```

```python
DATA_DIR = "./Hemoglobin PPG Dataset/Raw dataset per subject"
CLEAN_DIR = "./cleaned_ppg"

os.makedirs(CLEAN_DIR, exist_ok=True)

files = sorted([f for f in os.listdir(DATA_DIR) if f.endswith(".csv")])

print("Total subjects:", len(files))
print(files[:10])
```

```python
file_name = files[0]

path = os.path.join(DATA_DIR, file_name)

df = pd.read_csv(path)

print(df.head())
print("Shape:", df.shape)
```

```python
red = df["Red (a.u)"].values
ir  = df["Infra Red (a.u)"].values

age = df["Age"].iloc[0]
gender = df["Gender"].iloc[0]
hb = df["Hemoglobin (g/dL)"].iloc[0]

print("Age:", age)
print("Gender:", gender)
print("Hemoglobin:", hb)
```

```python
fs = 100

t = np.arange(len(red)) / fs
```

```python
end_s = 5
n = int(end_s * fs)

plt.figure(figsize=(10,3))

plt.plot(t[:n], red[:n], label="Red")
plt.plot(t[:n], ir[:n], label="Infrared")

plt.title(f"Raw PPG — Subject {file_name} (Hb={hb})")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

plt.show()
```

```python
def ppg_bandpass(sig, fs, low=0.5, high=5.0, order=2):

    nyq = fs/2

    b, a = signal.butter(order,
                         [low/nyq, high/nyq],
                         btype="bandpass")

    return signal.filtfilt(b, a, sig)
```

```python
red_clean = ppg_bandpass(red, fs)
ir_clean  = ppg_bandpass(ir, fs)
```

```python
plt.figure(figsize=(12,4))

plt.plot(t[:n], red[:n], label="Red raw", alpha=0.5)
plt.plot(t[:n], red_clean[:n], label="Red filtered")

plt.legend()
plt.title("PPG Cleaning — Raw vs Filtered")
plt.xlabel("Time (s)")
plt.grid(True)

plt.show()
```

```python
out_red = os.path.join(CLEAN_DIR,
                       file_name.replace(".csv","_red_clean.csv"))

out_ir  = os.path.join(CLEAN_DIR,
                       file_name.replace(".csv","_ir_clean.csv"))

np.savetxt(out_red, red_clean, fmt="%.6f")
np.savetxt(out_ir, ir_clean, fmt="%.6f")

print("Saved:", out_red)
print("Saved:", out_ir)
```

```python
for file in files:

    path = os.path.join(DATA_DIR, file)

    df = pd.read_csv(path)

    red = df["Red (a.u)"].values
    ir  = df["Infra Red (a.u)"].values

    red_clean = ppg_bandpass(red, fs)
    ir_clean  = ppg_bandpass(ir, fs)

    out_red = os.path.join(CLEAN_DIR,
                           file.replace(".csv","_red_clean.csv"))

    out_ir = os.path.join(CLEAN_DIR,
                          file.replace(".csv","_ir_clean.csv"))

    np.savetxt(out_red, red_clean, fmt="%.6f")
    np.savetxt(out_ir, ir_clean, fmt="%.6f")

    print("Cleaned:", file)
```

```python
CLEAN_DIR = "./cleaned_ppg"

red = np.loadtxt(os.path.join(CLEAN_DIR, "1_red_clean.csv"))
ir  = np.loadtxt(os.path.join(CLEAN_DIR, "1_ir_clean.csv"))

fs = 100

t = np.arange(len(red)) / fs
```

```python
from scipy.signal import find_peaks

peaks, _ = find_peaks(ir, distance=fs*0.4)

heart_rate = 60 / np.mean(np.diff(peaks)/fs)

print("Estimated Heart Rate:", heart_rate, "bpm")
```

```python
AC_red = red - np.mean(red)
AC_ir  = ir  - np.mean(ir)

DC_red = np.mean(red)
DC_ir  = np.mean(ir)

R = (np.std(AC_red)/DC_red) / (np.std(AC_ir)/DC_ir)

spo2 = 110 - 25*R

print("Estimated SpO2:", spo2, "%")
```

```python
from scipy import signal

def respiration_filter(sig, fs):

    b,a = signal.butter(2,
                        [0.1/(fs/2),0.4/(fs/2)],
                        btype="bandpass")

    return signal.filtfilt(b,a,sig)

resp = respiration_filter(ir, fs)
```

```python
from scipy.signal import find_peaks

resp_peaks,_ = find_peaks(resp, distance=fs*2)

resp_rate = 60 / np.mean(np.diff(resp_peaks)/fs)

print("Respiratory Rate:", resp_rate, "breaths/min")
```

```python
plt.figure(figsize=(12,8))

# PPG waveform
plt.subplot(4,1,1)
plt.plot(t[:1000], ir[:1000])
plt.plot(t[peaks][:20], ir[peaks][:20], "ro")
plt.title("PPG Waveform + Heartbeat Peaks")
plt.ylabel("Amplitude")
plt.grid(True)

# Heartbeat intervals
plt.subplot(4,1,2)
intervals = np.diff(peaks)/fs
plt.plot(intervals)
plt.title("Heartbeat Intervals")
plt.ylabel("Seconds")
plt.grid(True)

# Respiratory modulation
plt.subplot(4,1,3)
plt.plot(t[:2000], resp[:2000])
plt.title("Respiratory Signal (PPG Baseline Modulation)")
plt.ylabel("Amplitude")
plt.grid(True)

# SpO2 indicator
plt.subplot(4,1,4)
plt.bar(["SpO2"], [spo2])
plt.ylim(80,100)
plt.title("Estimated SpO₂")
plt.grid(True)

plt.tight_layout()
plt.show()
```
